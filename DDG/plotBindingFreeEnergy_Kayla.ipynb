```python
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import cm
import numpy as np
import pandas as pd
from collections import OrderedDict
```

```python
colors = cm.get_cmap('magma', 12)
```

```python
from pylab import *
print(matplotlib.colors.rgb2hex(colors(5/5)))
```

```python
#Raw data
mutations = ('S40D', 'S54A', 'A55D', 'D69Q', 'V71F', 'S54A', 'A55W', 'Y103F', 'S54A', 'S54A')
scores = np.concatenate((np.array([-1.27, 0.57, -0.83, -0.21, -0.36]), np.array([0.38, 0.52]), np.array([-0.71]), np.array([0.45]), np.array([0.47])))
antibodies = ["Pertuzumab"] * 5 + ["Atezolizumab"] * 2 + ["Daratumuab"] * 1 + ["Omalizumab"] * 1 + ["Trastzumab"] * 1
```

```python
d = {'Antibody': antibodies, 'Mutation': mutations, 'Score': scores}
df = pd.DataFrame(data=d)
```

```python
fig, ax = plt.subplots(1, 1, constrained_layout=True)

numTotAn = len(dict(zip(df['Antibody'], df['Antibody'])))
ht = ['+', '0', '\\', '*', '.']

antibody = []
numAn = 0
for index, row in df.iterrows():
    #define the mutation
    mutation = row['Mutation']
    
    #change the texture depending on the antibody
    if row['Antibody'] != antibody:
        antibody = row['Antibody']
        numAn += 1
    
    #change the color based on the mutation
    if "40" in mutation:
        cVal = 'blue'
    elif "55" in mutation and "54" not in mutation: 
        cVal = 'yellow'
    elif "54" in mutation and "55" not in mutation: 
        cVal = 'red'
    elif "55" and "54" in mutation: 
        cVal = 'orange'
    elif "69" in mutation and "71" not in mutation:
        cVal = 'green'
    elif "71" in mutation and "69" not in mutation: 
        cVal = 'pink'
    elif "69" and "71" in mutation: 
        cVal = '#A52A2A'
    elif "103" in mutation: 
        cVal = "purple"
    else:
        cVal = 'k'
    
    #plot the data
    ax.bar(index+1, row['Score'], color=cVal, label = mutation, hatch = ht[numAn-1])
    
#show the legend
plt.legend(bbox_to_anchor=(1, 1))

#axis labels
plt.xlabel("Antibody/Mutation")
plt.ylabel("$\Delta$$\Delta$G")

#show xticks
#plt.xticks(np.arange(1, df.shape[0]+1, 1), df['Antibody'], rotation = 90)
plt.xticks((3, 6.5, 8, 9, 10), ("Pertuzumab", "Atezolizumab", "Daratumuab", "Omalizumab", "Trastzumab"), rotation = 45)

#show the plot
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, constrained_layout=True)

plt.title("Impact of Framework Mutations on Free Energy")

per = np.array([-1.27, 0.57, -0.83, -0.21, -0.36])
xper = np.arange(1,len(per)+1,1)

ax.bar(xper, per, color=colors(1/5), label = "Pertuzumab") #ax.bar(x, height, width, bottom, align)

ate = np.array([0.38, 0.52])
xate = np.arange(len(xper)+1,len(xper)+len(ate)+1, 1)

ax.bar(xate, ate, color=colors(2/5), label = "Atezolizumab") #ax.bar(x, height, width, bottom, align)

dar = np.array([-0.71])
xdar = np.arange(len(xper)+len(xate)+1,len(xper)+len(xate)+len(dar)+1, 1)

ax.bar(xdar, dar, color=colors(3/5), label = "Daratumuab") #ax.bar(x, height, width, bottom, align)

oma = np.array([0.45])
xoma = np.arange(len(xper)+len(xate)+len(xdar)+1,len(xper)+len(xate)+len(xdar)+len(oma)+1, 1)

ax.bar(xoma, oma, color=colors(4/5), label = "Omalizumab") #ax.bar(x, height, width, bottom, align)

tra = np.array([0.47])
xtra = np.arange(len(xper)+len(xate)+len(xdar)+len(xoma)+1,len(xper)+len(xate)+len(xdar)+len(xoma)+len(tra)+1, 1)

ax.bar(xtra, tra, color=colors(5/5), label = "Trastzumab") #ax.bar(x, height, width, bottom, align)

ax.set_xlabel("Antibody/Mutation")
ax.set_ylabel("$\Delta$$\Delta$G")

ax.set_xticks(np.arange(1,len(xper) + len(xate) + len(xdar) + len(xoma) + len(xtra) + 1, 1))
#ax.set_xticklabels(('D40S', 'A54S', 'D55A', 'Q69D', 'F71V', 'A54S', 'W55A', 'F103Y', 'A54S', 'A54S'))
labels = ["Pertuzumab"] * len(xper) + ["Atezolizumab"] * len(xate) + ["Daratumuab"] * len(xdar) + ["Omalizumab"] * len(xoma) + ["Trastzumab"] * len(xtra)
ax.set_xticklabels(labels, rotation=40)

txt = "Interpreting the Results: \n Negative = Reversion was helpful (ie Mutation was not benefical) \n Positive = Reversion was destructive (ie Mutation was beneficial)"
plt.figtext(0.5, -0.2, txt, wrap=True, horizontalalignment='center', fontsize=12)
plt.legend(loc='lower right')

plt.show()
```
