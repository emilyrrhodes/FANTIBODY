```python
import matplotlib.pyplot as plt
from matplotlib import cm
import re
```

```python
filename = ['ate323_A55W_npt_prod_rmsd.xvg',
'ate323_control_npt_prod_rmsd.xvg',
'ate323_S54A_A55W_npt_prod_rmsd.xvg',
'ate323_S54A_npt_prod_rmsd.xvg',
'dar323_control_npt_prod_rmsd.xvg',
'dar323_Y103F_npt_prod_rmsd.xvg',
'oma366_control_npt_prod_rmsd.xvg',
'oma366_S54A_npt_prod_rmsd.xvg',
'per323_S40D_npt_prod_rmsd.xvg',
'per323_S54A_A55D_npt_prod_rmsd.xvg',
'per323_S54A_npt_prod_rmsd.xvg',
'per323_A55D_npt_prod_rmsd.xvg',
'per323_D69Q_V71F_npt_prod_rmsd.xvg',
'per323_D69Q_npt_prod_rmsd.xvg',
'per323_V71F_npt_prod_rmsd.xvg',
'per323_control_npt_prod_rmsd.xvg',
'tra366_control_npt_prod_rmsd.xvg',
'tra366_S54A_npt_prod_rmsd.xvg']
```

```python
colors = cm.get_cmap('magma', 12)
```

```python
##Individual graphs

for c in filename: 
    section = c[0:-4]
    name = 'results/' + section + '_200.xvg'
    with open(name, 'r') as file:
        title = ""
        xyaxis_labels = list()
        data_x = list()
        data_y = list()

        for line in file:
            if line.find("@") != -1:

                # Extracts the title
                if re.match("title", line.split()[1]):
                    title = line.split()[2].strip('\"')

                # Extracts the legends for the axes
                if len(line.split()) > 3:
                    if re.match("label", line.split()[2]):
                        xyaxis_labels.append(line.split('"')[1])

            if not line.startswith("@") and not line.startswith("#"):
                                  data_x.append(float(line.split()[0]))
                                  data_y.append(float(line.split()[1]))
                        
    val=c.index('_')
    antibody = c[:val]
    newC = c[val+1:]
    val2=newC.index('_')
    newC2 = newC[val2+1:]
    if newC2[0:8] != 'npt_prod':
        val3 = newC2.index('_')
        mutation = c[val+1:val+val2+val3+1]
    else:
        mutation = c[val+1:val+val2+1]
        
    if antibody == "ate323":
        antibody = "Atezolizumab"
    elif antibody == "dar323":
        antibody = "Daratumuab"
    elif antibody == "oma366":
        antibody = "Omalizumab"
    elif antibody == "per323":
        antibody = "Pertuzumab"
    elif antibody == "tra366":
        antibody = "Trastzumab"
        
    if "40" in mutation:
        cVal = 'blue'
    elif "55" in mutation and "54" not in mutation: 
        cVal = 'yellow'
    elif "54" in mutation and "55" not in mutation: 
        cVal = 'red'
    elif "55" and "54" in mutation: 
        cVal = 'orange'
    elif "69" in mutation and "71" not in mutation:
        cVal = 'green'
    elif "71" in mutation and "69" not in mutation: 
        cVal = 'pink'
    elif "69" and "71" in mutation: 
        cVal = '#A52A2A'
    elif "103" in mutation: 
        cVal = "purple"
    else:
        cVal = 'k'
        
    if "_" in mutation:
        unscInd = mutation.index('_')
        mutationFinal = mutation[:unscInd] + " & " + mutation[unscInd+1:]
    elif mutation == "control":
        mutationFinal = "Control"
    else: 
        mutationFinal = mutation
    
    fig, ax = plt.subplots(1, 1, constrained_layout=True)
    if cVal == colors(5/5):
        ax.plot(data_x, data_y, color = "k", linewidth = 2)
        ax.plot(data_x, data_y, color = cVal, linewidth = 1)
    else: 
        ax.plot(data_x, data_y, color = cVal, linewidth = 2)
    ax.set_ylim([0,16])
    ax.set_xlabel("Time [ps]")
    ax.set_ylabel("Root Mean Square Deviation [nm]")
        
    
    title = antibody + ": " + mutationFinal
    
    ax.set_title(title)
    fig.show()
    
    #print(c[0:-4] + ".jpg")
    saveName = c[0:-4] + "_200.jpg"
    
    plt.savefig(saveName)
```

```python
antibodyNames = ["Atezolizumab", "Daratumuab", "Omalizumab", "Pertuzumab", "Trastzumab"]

for x in antibodyNames: 
    if x == "Atezolizumab":
        files = ['ate323_A55W_npt_prod_rmsd.xvg',
'ate323_control_npt_prod_rmsd.xvg',
'ate323_S54A_A55W_npt_prod_rmsd.xvg',
'ate323_S54A_npt_prod_rmsd.xvg']
        plotRMSD(files)
    elif x == "Daratumuab":
        files = ['dar323_control_npt_prod_rmsd.xvg',
'dar323_Y103F_npt_prod_rmsd.xvg']
        plotRMSD(files)
    elif x == "Omalizumab":
        files = ['oma366_control_npt_prod_rmsd.xvg',
'oma366_S54A_npt_prod_rmsd.xvg']
        plotRMSD(files)
    elif x == "Pertuzumab":
        files = ['per323_S40D_npt_prod_rmsd.xvg',
'per323_S54A_A55D_npt_prod_rmsd.xvg',
'per323_S54A_npt_prod_rmsd.xvg',
'per323_A55D_npt_prod_rmsd.xvg',
'per323_D69Q_V71F_npt_prod_rmsd.xvg',
'per323_D69Q_npt_prod_rmsd.xvg',
'per323_V71F_npt_prod_rmsd.xvg',
'per323_control_npt_prod_rmsd.xvg']
        plotRMSD(files)
    elif x == "Trastzumab":
        files = ['tra366_control_npt_prod_rmsd.xvg',
'tra366_S54A_npt_prod_rmsd.xvg']
        plotRMSD(files)  
```

```python
def plotRMSD(files):
    fig, ax = plt.subplots(1, 1, constrained_layout=True)
    numFiles = len(files)
    fileNum = 0
    for c in files: 
        fileNum += 1
        section = c[0:-4]
        name = 'results/' + section + '_200.xvg'
        with open(name, 'r') as file:
            title = ""
            xyaxis_labels = list()
            data_x = list()
            data_y = list()

            for line in file:
                if line.find("@") != -1:

                    # Extracts the title
                    if re.match("title", line.split()[1]):
                        title = line.split()[2].strip('\"')

                    # Extracts the legends for the axes
                    if len(line.split()) > 3:
                        if re.match("label", line.split()[2]):
                            xyaxis_labels.append(line.split('"')[1])

                if not line.startswith("@") and not line.startswith("#"):
                                      data_x.append(float(line.split()[0]))
                                      data_y.append(float(line.split()[1]))

        mutation = ''
        val=c.index('_')
        antibody = c[:val]
        newC = c[val+1:]
        val2=newC.index('_')
        newC2 = newC[val2+1:]
        if newC2[0:8] != 'npt_prod':
            val3 = newC2.index('_')
            mutation = c[val+1:val+val2+val3+1+1]
        else:
            mutation = c[val+1:val+val2+1]

        if "40" in mutation:
            cVal = 'blue'
        elif "55" in mutation and "54" not in mutation: 
            cVal = 'yellow'
        elif "54" in mutation and "55" not in mutation: 
            cVal = 'red'
        elif "55" and "54" in mutation: 
            cVal = 'orange'
        elif "69" in mutation and "71" not in mutation:
            cVal = 'green'
        elif "71" in mutation and "69" not in mutation: 
            cVal = 'pink'
        elif "69" and "71" in mutation: 
            cVal = '#A52A2A'
        elif "103" in mutation: 
            cVal = "purple"
        else:
            cVal = 'k'
            
           
        if "_" in mutation:
            unscInd = mutation.index('_')
            mutationFinal = mutation[:unscInd] + " & " + mutation[unscInd+1:]
        elif mutation == "control":
            mutationFinal = "Control"
        else: 
            mutationFinal = mutation
            
        name = antibody + ": " + mutationFinal
        
        if mutation == "control":
            ax.plot(data_x, data_y, color = "k", linewidth = 2, alpha = 1 ,label = mutationFinal)
        elif cVal == colors(5/5):
            ax.plot(data_x, data_y, color = "k", linewidth = 2)
            ax.plot(data_x, data_y, color = cVal, linewidth = 1, label = mutationFinal)
        else: 
            ax.plot(data_x, data_y, color = cVal, linewidth = 2, label = mutationFinal)
        ax.set_ylim([0,16])
        ax.set_xlabel("Time [ps]")
        ax.set_ylabel("Root Mean Square Deviation [nm]")
        ax.legend(loc="upper left")

        if antibody == "ate323":
            antibody = "Atezolizumab"
        elif antibody == "dar323":
            antibody = "Daratumuab"
        elif antibody == "oma366":
            antibody = "Omalizumab"
        elif antibody == "per323":
            antibody = "Pertuzumab"
        elif antibody == "tra366":
            antibody = "Trastzumab"
        
        title = antibody
        
        ax.set_title(title)
        fig.show()

    saveName = antibody + "_200.jpg"
    plt.savefig(saveName)
        
```
